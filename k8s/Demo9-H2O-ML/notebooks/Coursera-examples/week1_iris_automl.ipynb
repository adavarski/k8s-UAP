# Week 1 - Iris Auto ML

Now let's look at this again, but using H2O's Auto ML feature.

```python
import h2o
```

```python
# check to see if H2O is running already and connect, or start it up
h2o.init(ip='h2o-service')
```

```python
# import the iris dataset
url = 'http://h2o-public-test-data.s3.amazonaws.com/smalldata/iris/iris_wheader.csv'
iris = h2o.import_file(url)
```

```python
# split the data into 80% train and 20% test
train, test = iris.split_frame([0.8])
```

```python
# review a summary of our training data
train.summary()
```

```python
# how many rows in training - not exactly 80%
train.nrows
```

```python
# how many rows in testing - not exactly 20%
test.nrows
```

```python
# import auto ml
from h2o.automl import H2OAutoML
```

```python
# train the model
mA = H2OAutoML(max_runtime_secs=60)
mA.train(['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid'], 'class', train)
```

```python
# view the leader model
mA.leader
```

```python
# predict on the test set
p = mA.leader.predict(test)
```

```python
# review predictions
p
```

```python
# review model performance on the test set
mA.leader.model_performance(test)
```
