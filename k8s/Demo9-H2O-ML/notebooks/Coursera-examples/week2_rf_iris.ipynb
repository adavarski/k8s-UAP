# Week 3 - Random Forest with the Iris Dataset

```python
import h2o
```

```python
# check to see if H2O is running already and connect, or start it up
h2o.init()
```

```python
# import the iris dataset
url = 'http://h2o-public-test-data.s3.amazonaws.com/smalldata/iris/iris_wheader.csv'
iris = h2o.import_file(url)
```

```python
# split the data into 80% train and 20% test
train, test = iris.split_frame([0.8])
```

```python
# review a summary of our training data
train.summary()
```

```python
train.nrows
```

```python
test.nrows
```

```python
# import the RF estimator
from h2o.estimators.random_forest import H2ORandomForestEstimator
```

```python
mRF = H2ORandomForestEstimator()
mRF.train(['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid'], 'class', train)
```

```python
# see the model results
mRF
```

```python
# see results on the test set
mRF.model_performance(test)
```
