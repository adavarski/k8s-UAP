# Week 4 - Deep Learning Assignment

```python
# import packages
import h2o
```

```python
# used 3 cores and 1.520 Gb memory
h2o.init()
```

```python
# import the cacao data
data = h2o.import_file('http://coursera.h2o.ai/cacao.882.csv')
```

```python
# split the data
train, valid, test = data.split_frame([0.8,0.1], seed=69)
```

```python
# for the classification problem - y is 'Maker Location'
y = 'Maker Location'
ignore_fields = ['Maker Location']
x_all = [i for i in train.names if i not in ignore_fields]
```

```python
# import the deep learning estimator
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
```

```python
# instantiate the default model and time the training, added seed for reproducability
# runs in about 9 seconds
m_def = H2ODeepLearningEstimator(seed=69)
%time m_def.train(x_all, y, train, validation_frame=valid)
```

```python
# show logloss values for train and valid for default model
m_def.logloss(train=True, valid=True)
```

```python
# review performance on test data
m_def.model_performance(test)
```

```python
# add more layers, add dropout, increase epochs, and add regularization
# the default early stopping plus dropout and regularization should ensure we do not overfit
# runs in about 23 seconds, which is less than 10 times the runtime of the default model
m_tuned = H2ODeepLearningEstimator(
                activation = 'RectifierWithDropout',
                epochs=40,
                hidden=[200,200,200],
                hidden_dropout_ratios=[0.2, 0.2, 0.2],
                l1=1e-5,
                l2=1e-5,
                seed=69
)
%time m_tuned.train(x_all, y, train, validation_frame=valid)
```

```python
# show logloss values for train and valid for tuned model
m_tuned.logloss(train=True, valid=True)
```

```python
# review performance on test data
m_tuned.model_performance(test)
```

## Summary

In comparing the default model to the tuned model, the logloss has decreased on each of train, valid, and test. Keeping the early stopping, as well as adding dropout and regularization, should ensure that the model is not overfitting.

```python
# save the default model
h2o.save_model(model=m_def, force=True)
```

```python
# save the tuned model
h2o.save_model(model=m_tuned, force=True)
```
