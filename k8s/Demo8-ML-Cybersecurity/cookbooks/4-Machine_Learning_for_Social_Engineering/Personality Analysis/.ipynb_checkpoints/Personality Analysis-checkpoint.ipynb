```python
from ibm_watson import PersonalityInsightsV3
from datetime import date
v = str(date.today())
api_key ="fill me in"
```

```python
personalityInsightsService = PersonalityInsightsV3(
    version=v,
    iam_apikey=api_key)
```

```python
tweetsFile = "ElonMuskTweets.txt"
```

```python
with open(tweetsFile) as inputFile:
    profile = personalityInsightsService.profile(
        inputFile.read(),
        'application/json',
        raw_scores=False,
        consumption_preferences=True).get_result()
```

```python
import json
print(json.dumps(profile, indent=2))
```
